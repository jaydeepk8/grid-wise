```python
import joblib
import pandas as pd
from sklearn.ensemble import RandomForestRegressor


df = pd.read_csv("../data/hospital_ml_ready.csv")

X = df[
    [
        'hour',
        'day_of_week',
        'is_weekend',
        'prev_hour_energy',
        'rolling_3h_avg',
        'rolling_6h_avg'
    ]
]

y = df['target_next_hour']


rf_model = RandomForestRegressor(
    n_estimators=200,
    max_depth=12,
    random_state=42,
    n_jobs=-1
)

rf_model.fit(X, y)

# Save model
joblib.dump(rf_model, "../model/hospital_energy_rf.pkl")

print(" Model saved successfully")
```
